```python
#nested if else
num = 10

if num > 0:
    if num % 2 == 0:
        print("Positive Even number")
    else:
        print("Positive Odd number")
else:
    print("Negative number")
```

```python
#break
for i in range(1, 6):
    if i == 3:
        break
    print(i)
```

```python
# match case
day = 2

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case _:
        print("Invalid day")
```

```python
#ternary if else
n = 7
result = "Even" if n % 2 == 0 else "Odd"
print(result)
```
